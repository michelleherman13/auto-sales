```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json
```

```python
df = pd.read_csv('final_data.csv')
df
```

```python
five_years_df=df[df["Year"]>2016]
five_years_df.head()
```

```python
fig, ax = plt.subplots(figsize=(5,5))

x_values = five_years_df['Price Per Gallon']
y_values = five_years_df['Sales']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.title('2017-2021 LR')
plt.xlabel('Price Per Gallon')
plt.ylabel('Sales (1,000,000)')
plt.grid(linestyle="-", linewidth=1, alpha=0.25)
print(f"The r-squared is: {rvalue**2}")
plt.show()

fig.savefig('Images/2017-2021 Sales vs PPG LR.png', dpi=100)
```
