```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
data_df = pd.read_csv("final_data.csv")
```

```python
data_df
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.title("Sales and Gas Price")

ax2=ax.twinx()
ax.plot(data_df["MonthYear"], data_df["Sales"], color="blue", label="Sales" )
ax2.plot(data_df["MonthYear"], data_df["Price Per Gallon"], 
                        color="green",label="Gas Price")

ax.set_xlabel("2005-2021")
ax.set_ylabel("Vehicles Sold")
ax2.set_ylabel("Gas Price")

# Create a legend for our chart
ax2.legend(loc=0)

fig.savefig('Images/2005-2021 Sales and Gas Price.png', dpi=100)

# Show the chart
plt.tight_layout()
plt.show()
```
